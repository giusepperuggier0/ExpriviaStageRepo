### NumPy

La libreria **NumPy** (Numerical Python) è l'implementazione Python preferita per lavorare con gli Array.
- Offre un tipo di array n-dimensionale ad alte prestazioni chiamato `ndarray`
- Le operazioni sugli array sono fino a due ordini di grandezza più veloci di quelle sulle liste.
- Oltre $450$ librerie Python dipendono da NumPy. Molte popolari librerie di data science come Pandas, SciPy (Scientific Python) e Keras (per il deep learning) sono basate o dipendono da NumPy.

Il punto di forza di NumPy è la ''programmazione orientata agli array'', che utilizza la programmazione in stile funzionale con iterazione interna per rendere le manipolazioni degli array concise e dirette, eliminando i tipi di bug che possono verificarsi con l'iterazione esterna di cicli programmati esplicitamente.

````python
import numpy as np

x = np.arange(15)  # range di valori da 0 a 14

print(x)
print(type(x))     # questo oggetto non è una lista ma un array numpy
````

````python
x = np.arange(15)  # range di valori da 0 a 14
y = np.arange(15,0,-1)  # range di valori da 15 a 0

print('x:', x)
print('y:', y)

# Alcuni metodi e operazioni per array numpy
print('Metodi per array:')
z = x.copy()       # Copia del vettore, simile a quanto visto con le liste
print('z:', z)
print(z.argmax())  # indice del valore massimo
print(z.max())     # valore massimo nell'array
print(z.argmin())  # indice del valore massimo
print(z.min())     # valore minimo nell'array
print(z.reshape(5,3))  # Reshape del vettore in altra forma (es. matrice 5X3)
print(z.reshape(5,3).ravel())  # Flattizzare una matrice
print(z.sum())     # Somma di tutti gli elementi dell'array
print(z.reshape(5,3).transpose()) # Trasposizione di matrice
print(z.reshape(5,3).T) # Trasposizione di matrice
np.shape(x)  # Forma di x
print(np.concatenate((x, y), axis=0)) # Concatenazione di array
````

#### Disegnare un Plot da uno script

Se stai usando Matplotlib dall'interno di uno script, la funzione ``plt.show()`` avvia un ciclo di eventi, cerca tutti gli oggetti figura attualmente attivi e apre una o più finestre interattive che mostrano la tua figura o le tue figure.

Quindi, ad esempio, potresti avere un file chiamato *myplot.py* contenente quanto segue:

```python
# ------- file: myplot.py ------
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

È quindi possibile eseguire questo script dal prompt della riga di comando, che si tradurrà in una finestra che si apre con la figura visualizzata:

```
$ python myplot.py
```

Il comando ``plt.show()`` nasconde molto bene i dettagli implementativi, poiché deve interagire con il backend grafico interattivo del tuo sistema.
I dettagli di questa operazione possono variare notevolmente da sistema a sistema e persino da installazione a installazione.

#### Disegnare un Plot da un IPython notebook

Il notebook IPython è uno strumento di analisi dei dati interattivo basato su browser che può combinare narrativa, codice, grafica, elementi HTML e molto altro in un unico documento eseguibile.

La visualizzazione interattiva all'interno di un notebook IPython può essere eseguita con il comando ``% matplotlib`` e funziona in modo simile alla shell IPython.
Nel notebook IPython, hai anche la possibilità di incorporare la grafica direttamente nel notebook, con due possibili opzioni:

- ``%matplotlib notebook`` porterà a grafici *interattivi* incorporati nel notebook
- ``%matplotlib inline`` porterà a grafici *statici*  del tuo plot incorporati nel notebook

In genere opteremo per ``%matplotlib inline``:

````python
%matplotlib inline
````

Dopo aver eseguito questo comando (deve essere eseguito solo una volta per kernel / sessione), qualsiasi cella all'interno del notebook che crea un grafico incorporerà un'immagine PNG della grafica risultante

Ora immaginate di voler scoprire la valutazione media, non per ogni singola app, ma piuttosto per ogni singolo genere di app. 

Possiamo usare un semplice "groupby", passando il nome della colonna con cui vogliamo raggruppare come primo argomento, e anche impostando una "funzione di aggregazione" - nel nostro caso sarà ``np.mean`` come vogliamo valore medio. 

Infine, una volta eseguito il metodo groupby, estraiamo semplicemente la colonna che ci interessa utilizzando la notazione delle parentesi standard alla fine per selezionare quella colonna.

````python
genre_rating = df.groupby('prime_genre').mean(numeric_only=True)['user_rating']
genre_rating
````

È davvero molto semplice quindi andare e tracciare un semplice grafico a barre usando matplotlib

````python
plt.bar(x=genre_rating.index, height=genre_rating)
plt.show();
````

````python
# set size of overall figure
plt.figure(figsize=(20,10))
plt.bar(x=genre_rating.index, height=genre_rating)
plt.show();
````

````python
genre_rating = genre_rating.sort_values()

# set size of overall figure
plt.figure(figsize=(20,10))
plt.barh(y=genre_rating.index, width=genre_rating)
plt.show();
````

````python
# set size of overall figure
plt.figure(figsize=(20,10))

# set chart title
plt.title('Average app User Rating by Genre')

# set x-axis label
plt.xlabel('User Rating')

# set y-axis label
plt.ylabel('Genre')

plt.barh(y=genre_rating.index, width=genre_rating)

plt.show();
````

````python
# reset styles to default
plt.rcdefaults()

# set new style
plt.style.use('dark_background')

# set size of overall figure
plt.figure(figsize=(20,10))

# set chart title
plt.title('Average app User Rating by Genre')

# set x-axis label
plt.xlabel('User Rating')

# set y-axis label
plt.ylabel('Genre')

plt.barh(y=genre_rating.index, width=genre_rating)

plt.show();
````

Possiamo persino modellare la figura da soli se non siamo soddisfatti degli stili incorporati. Ovviamente ci vuole più tempo per farlo in questo modo, ma matplotlib ti permette davvero di controllare ogni piccola minuzia che potresti sperare di modellare.

Diamo un'occhiata solo agli ultimi 10 generi in base alla valutazione media e questa volta iniziamo una nuova trama con uno stile personalizzato: vediamo cosa possiamo inventare.

Quindi per prima cosa estraiamo solo gli ultimi 10 generi. Potremmo farlo con un semplice approccio slice poiché abbiamo già ordinato i nostri dati per dimensione, tuttavia per essere più esplicito userò il metodo ``nsmallest()`` che seleziona gli "n" punti dati più piccoli, dove "n" è il numero passato al metodo come argomento. 

Quindi nel nostro caso, per estrarre gli ultimi 10 useremo:

Di seguito codifichiamo il grafico effettivo: implica la compressione e l'iterazione dei dati user_rating per creare le etichette dei punti dati e così via. 
I bordi del plot vengono quindi schiariti e vengono aggiunti un titolo e un'etichetta x.
